## (optional) The `enumerate()` function

In cases where the list element **and** a numerical index is required, Python provides the `enumerate()` function to allow the use of `for` loop without requiring `range()` function.

Try the following code cell below to understand how `enumerate()` works:

```python
fruit = ['apple','banana','cherry','durian','elderberry']

for each_item in enumerate(fruit):
    print(each_item)
```

Notice that `each_item` is a pair of variables, consisting of an `int` and a `str`.

You can store them into two separate variables by declaring two variables in the `for` loop:  
(Run the code cell below.)

```python
for i,item in enumerate(fruit):
    print(f'Item {i}: {item}')
```
